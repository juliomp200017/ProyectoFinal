```python
import os
import mysql.connector
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
user_name = os.environ.get('DB_USER')
password = os.environ.get('DB_PASS')
```

```python
connection = mysql.connector.connect(host='mysql',
                                     user='root',
                                     password='12345678',
                                     db='Datos')
```

```python
connection
```

```python
datos_tables1=pd.read_sql_query('SELECT * FROM casos', connection)
```

```python
datos_tables1
```

```python
y_pos = np.arange(len(datos_tables1.municipio))
 
plt.figure(figsize=(15,10))
plt.barh(y_pos, datos_tables1.numcasos, align='center', alpha=0.5)
plt.yticks(y_pos, datos_tables1.municipio)
plt.xlabel('Casos')
plt.title('# De casos positivos en Risaralda')

plt.show()
```

```python
datos_tables2=pd.read_sql_query('SELECT * FROM CasosAtla', connection)
```
